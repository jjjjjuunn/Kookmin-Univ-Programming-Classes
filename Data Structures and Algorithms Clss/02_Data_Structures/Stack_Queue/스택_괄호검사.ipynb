```python
# ì—°ìŠµë¬¸ì œ #5-1 : ì˜¬ë°”ë¥¸ ê´„í˜¸ (programmers) ìŠ¤íƒ/í Lv.2 - 12909
# (ë¬¸ì œ) ì—¬ëŸ¬ë¶„ì´ í’€ì–´ì•¼ í•  ë¬¸ì œëŠ” ì£¼ì–´ì§„ ê´„í˜¸ì—´ì„ ì½ê³ , ê´„í˜¸ê°€ ì˜¬ë°”ë¥´ê²Œ ì§ì§€ì–´ì¡ŒëŠ”ì§€ ì—¬ë¶€ë¥¼ ì¶œë ¥í•˜ëŠ” ê²ƒì´ë‹¤. 
def solution(s):
    stack = []
    for ch in s:
        if ch == '(': stack.append(ch)
        else:
            if stack: stack.pop()
            else: return False
    return not stack

s = input()
print(solution(s))
```

```python
# ì—°ìŠµë¬¸ì œ 5-1: ê´„í˜¸ í™•ì¥í•˜ì—¬ ì‘ì„±í•˜ê¸° ({[]])

def testPair(exp:str):
    s = []
    for st in exp:
        if st == '(' or st == '{' or st == '[':
            s.append(st)
        elif st == ')' or st == '}' or st == ']':
            if not len(s):
                return False
            symbol = s[-1]
            s.pop()
            if st == ')' and symbol != '(':
                return False
            elif st == '}' and symbol != '{':
                return False
            elif st == ']' and symbol != '[':
                return False
            
    if not len(s):
        return True
    
    return False

exp = list(input())
print(testPair(exp))
```

```python
# isOperator

def isOperator(op: str)-> bool:
    return op == '+' or op == '-' or op == '*' or op == '/'

def precedence(op: str)-> int:
    if op == '(' : return 0
    elif op == '+' or op == '-': return 1
    elif op == '*' or op == '/': return 2
    else: return 3
```

```python
# InfixToPostfix()

def InfixToPostfix(infix: str)-> str:
    s = []

    infix = infix.replace('(', ' ( ')
    infix = infix.replace(')', ' ) ')
    infix = infix.replace('+', ' + ')
    infix = infix.replace('-', ' - ')
    infix = infix.replace('*', ' * ')
    infix = infix.replace('/', ' / ')

    infix = infix.split()

    postfix = []
    for ch in infix:
        if ch == '(':
            s.append(ch)

        elif ch == ')':
            while s[-1] != '(':
                postfix.append(s[-1])
                s.pop()
            s.pop()
        
        elif isOperator(ch):
            while len(s) and precedence(s[-1]) >= precedence(ch):
                postfix.append(s[-1])
                s.pop()
            s.append(ch)
        
        elif str.isdigit(ch):
            postfix.append(ch)

        elif ch == ' ': 
            continue
        else:
            print('ì˜ëª»ëœ ìˆ˜ì‹ !!!')
            return
            
    while s:
        postfix.append(s[-1])
        s.pop()
    return postfix
```

```python
def evalPostfix(postfix: str)-> int:
    s = []
    for ch in postfix:
        if str.isdigit(ch):
            s.append(ch)

        elif isOperator(ch):
            op2 = int(s[-1])
            s.pop()
            op1 = int(s[-1])
            s.pop()
            if ch == '+':
                s.append(op1 + op2)
            elif ch == '-':
                s.append(op1 - op2)
            elif ch == '*':
                s.append(op1 * op2)
            elif ch ==  '/':
                s.append(op1 / op2)
            elif ch == ' ':
                continue
            else:
                print('ì˜ëª»ëœ ìˆ˜ì‹!!!')
                return
    if s:
        res = s[-1]
        s.pop()
    return res
    
```

```python
#main
if __name__ == '__main__':
    infix = input('ìˆ˜ì‹ ì…ë ¥: ')
    
    postfix = InfixToPostfix(infix)

    print(f'ìˆ˜ì‹ ì…ë ¥: {infix}')
    print(f'í›„ìœ„í‘œê¸°ë²• ë³€í™˜: {postfix}')
    
    res = evalPostfix(postfix)
    print(f'ì—°ì‚°ê²°ê³¼: {res}')
```
