```python
# HeapSort: í™ ì •ë ¬
def heapSort(sList) -> None:
    buildHeap(sList)
    for last in range(len(sList) -1, 0, -1):
        sList[last], sList[0] = sList[0], sList[last]
        downHeap(sList, 0, last-1)
    # deleteMax();
    # downHeap();

# buildHeap: ìµœëŒ€ í™ êµ¬ì„±
def buildHeap(sList) -> None:
    for i in range((len(sList)-1) // 2, -1, -1):
        downHeap(sList, i, len(sList)-1)

# downHeap: ìµœëŒ€ í™ìœ¼ë¡œ ì¬êµ¬ì„±
def downHeap(sList, k:int, end:int) ->None:
    child = 2 * k + 1
    right = 2 * k + 2
    if child <= end:
        if right <= end and sList[child] < sList[right]:
            child = right
        if sList[k] < sList[child]:
            sList[k], sList[child] = sList[child], sList[k]
            downHeap(sList, child, end)

if __name__ == '__main__' :
    import random

    # sList = [5,2,3,9,6,1,8,4,7]
    sList = []
    while len(sList)<15:
        num = random.randint(0, 99)
        sList.append(num)
    
    print(f'ì •ë ¬ ì „: {sList}')
    heapSort(sList)
    print(f'ì •ë ¬ í›„: {sList}')
```

```python
'''
	ì´ì§„ ê²€ìƒ‰ íŠ¸ë¦¬: ì•Œê³ ë¦¬ì¦˜ êµ¬í˜„(Python) -- ì´ì¤‘ì—°ê²°ë¦¬ìŠ¤íŠ¸
		íŒŒì¼ëª…: LinkedBSTree.py
        	- __main__: ë°ì´í„° ì‚½ì….ì‚­ì œ, ì „ì²´ ì›ì†Œ ì¶œë ¥
			- í´ë˜ìŠ¤: DNode
			- í´ë˜ìŠ¤: LinkedBSTree
				ë°ì´í„° ì‚½ì…: InsertBST(_insertBST)
				ë°ì´í„° ì‚­ì œ: DeleteBST(_deleteBST)
				ë°ì´í„° ê²€ìƒ‰: SearchBST(_searchBST)
				ì „ì²´ ì›ì†Œ ì¶œë ¥: printBSTAll(Preorder)
'''

# í´ë˜ìŠ¤ ì„¤ê³„: LinkedBSTree
class LinkedBSTree:
    class DNode:
        def __init__ (self, data, Llink=None, Rlink=None):
            self.data = data
            self.Llink = Llink
            self.Rlink = Rlink

    def __init__(self):
        self.__root = None

    # ì†Œë©¸ì: ë„ˆë¹„ ìš°ì„  ìˆœíšŒ
    def __del__(self):
        if not self.__root : return
			
    # ì „ì²´ ë…¸ë“œ ì¶œë ¥: ë„ˆë¹„ ìš°ì„  ìˆœíšŒ
    def printBSTAll(self) -> None:
        if not self.__root:
            print("\në°ì´í„°ê°€ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
            return

    # ë°ì´í„° ê²€ìƒ‰: ì¬ê·€ì .ë¹„ì¬ê·€ì  ìš©ë²•
    def searchBST(self) -> None:
        # ë°ì´í„° ê²€ìƒ‰: ì¬ê·€ì  ìš©ë²•,  ë¹„ì¬ê·€ì  ìš©ë²•
        def _searchBST(root, data):
            if root == None or data == root.data:
                return root
            elif data < root.data :
                return _searchBST(root.Llink, data)
            else:
                return _searchBST(root.Rlink, data)

        if not self.__root:
            print("\në°ì´í„°ê°€ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
            return
        
        print("\nì´ì§„ ê²€ìƒ‰ íŠ¸ë¦¬: ë°ì´í„° ê²€ìƒ‰")
        while True:
            num = int(input('ì„ì˜ì˜ ì •ìˆ˜ ì…ë ¥(ì¢…ë£Œ: 0): '))
            if num == 0:
                break
            tNode = _searchBST(self.__root, num)
            if tNode:   print(f'{tNode.data} í‚¤ë¥¼ ì°¾ì•˜ìŠµë‹ˆë‹¤!!!')
            else:       print('í‚¤ë¥¼ ì°¾ì§€ ëª»í–ˆìŠµë‹ˆë‹¤.')

    # ë°ì´í„° ì‚½ì…: ì¬ê·€ì .ë¹„ì¬ê·€ì  ìš©ë²•
    def insertBST(self) -> None:
        # ë°ì´í„° ì‚½ì…: ì¬ê·€ì  ìš©ë²•, ë¹„ì¬ê·€ì  ìš©ë²•
        def _insertBST(root, data):
            if root == None:
                root = self.DNode(data)
            elif data < root.data:
                root.Llink = _insertBST(root.Llink, data)
            else:
                root.Rlink = _insertBST(root.Rlink, data)
            return root

        print("\nì´ì§„ ê²€ìƒ‰ íŠ¸ë¦¬: ë°ì´í„° ì‚½ì…")
        while True:
            num = int(input('ì„ì˜ì˜ ì •ìˆ˜ ì…ë ¥(ì¢…ë£Œ: 0): '))
            if num == 0:
                break
            self.__root = _insertBST(self.__root, num)

    # ë°ì´í„° ì‚­ì œ: ì¬ê·€ì .ë¹„ì¬ê·€ì  ìš©ë²•
    def deleteBST(self) -> None:
        # ë°ì´í„° ì‚­ì œ: ë¹„ì¬ê·€ì  ìš©ë²•
        def _deleteBST(root, data):
            if not self.__root:
                print("\në°ì´í„°ê°€ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
                return
            pParent = None
            tNode = root

            while data != tNode.data:
                pParent = tNode
                if data < tNode.data:
                    tNode = tNode.Llink
                else:
                    tNode = tNode.Rlink

            if tNode.Llink == None and tNode.Rlink == None:     # ìì‹ ë…¸ë“œê°€ ì—†ëŠ” ê²½ìš°
                if pParent:
                    if pParent.Llink == tNode:
                        pParent.Llink = None
                    else:
                        pParent.Rlink = None
                else:
                    root = None

            # Case 2: ìì‹ ë…¸ë“œê°€ 1ê°œ ìˆëŠ” ê²½ìš°
            elif tNode.Llink == None or tNode.Rlink == None:
                # pChild: ì‚­ì œí•  ë…¸ë“œì˜ ìì‹ ë…¸ë“œ
                if tNode.Llink:
                    pChild= tNode.Llink
                else:
                    pChild = tNode.Rlink

                if pParent:
                    if pParent.Llink == tNode:
                        pParent.Llink = pChild
                    else:
                        pParent.Rlink = pChild
                else:
                    root = pChild

            # Case 3: ìì‹ ë…¸ë“œê°€ ë‘˜ ë‹¤ ìˆëŠ” ê²½ìš°
            elif tNode.Llink != None and tNode.Rlink != None:
                # ì™¼ìª½ ì„œë¸Œ íŠ¸ë¦¬ì—ì„œ í›„ê³„ì(ìµœëŒ“ê°’) ì°¾ê¸°
                sParent = tNode
                sNode = tNode.Llink
                while sNode.Rlink:
                    sParent = sNode
                    sNode = sNode.Rlink

                if sParent.Llink == sNode:
                    sParent.Llink = sNode.Llink
                else:
                    sParent.Rlink = sNode.Llink

                tNode.data = sNode.data
                tNode = sNode

            del tNode
            return root 

        print("\nì´ì§„ ê²€ìƒ‰ íŠ¸ë¦¬: ë°ì´í„° ì‚­ì œ")
        while True:
            num = int(input('ì„ì˜ì˜ ì •ìˆ˜ ì…ë ¥(ì¢…ë£Œ: 0): '))
            if num == 0:
                break
            self.__root = _deleteBST(self.__root, num)

if __name__ == '__main__':
    import os	# system
    import sys  # exit

    bst = LinkedBSTree()
    while True:
        os.system('clear')
        print('\n##### ì´ì§„ ê²€ìƒ‰ íŠ¸ë¦¬ ###')
        print('1) ë°ì´í„° ì‚½ì…')
        print('2) ë°ì´í„° ì‚­ì œ')
        print('3) ë°ì´í„° ê²€ìƒ‰')
        print('4) ì „ì²´ ì¶œë ¥')
        print('5) í”„ë¡œê·¸ë¨ ì¢…ë£Œ\n')
        choice = int(input('ë©”ë‰´ ì„ íƒ: '))

        match choice:
            case 1: bst.insertBST()
            case 2: bst.deleteBST()
            case 3: bst.searchBST()
            case 4: bst.printBSTAll()
            case 5:
                sys.exit("\ní”„ë¡œê·¸ë¨ ì¢…ë£Œ!!!")
            case _: print('\nì˜ëª» ì„ íƒ í•˜ì…¨ìŠµë‹ˆë‹¤. \n')
        input("ê³„ì†í•˜ë ¤ë©´ ì—”í„°í‚¤ë¥¼ ì…ë ¥í•˜ì‹­ì‹œì˜¤.")

    # del bst
    # bst.__del__
```
