# 힙 정렬 구현해보기(Python)

```python
# HeapSort: 힙 정렬
def heapSort(sList) -> None:
    buildHeap(sList)
    for last in range(len(sList) -1, 0, -1):
        sList[last], sList[0] = sList[0], sList[last]
        downHeap(sList, 0, last-1)
    # deleteMax();
    # downHeap();

# buildHeap: 최대 힙 구성
def buildHeap(sList) -> None:
    for i in range((len(sList)-1) // 2, -1, -1):
        downHeap(sList, i, len(sList)-1)

# downHeap: 최대 힙으로 재구성
def downHeap(sList, k:int, end:int) ->None:
    child = 2 * k + 1
    right = 2 * k + 2
    if child <= end:
        if right <= end and sList[child] < sList[right]:
            child = right
        if sList[k] < sList[child]:
            sList[k], sList[child] = sList[child], sList[k]
            downHeap(sList, child, end)

if __name__ == '__main__' :
    import random

    # sList = [5,2,3,9,6,1,8,4,7]
    sList = []
    while len(sList)<15:
        num = random.randint(0, 99)
        sList.append(num)
    
    print(f'정렬 전: {sList}')
    heapSort(sList)
    print(f'정렬 후: {sList}')
```

## 재귀적 용법으로 구현해보기

```python
'''
	이진 검색 트리: 알고리즘 구현(Python) -- 이중연결리스트
		파일명: LinkedBSTree.py
        	- __main__: 데이터 삽입.삭제, 전체 원소 출력
			- 클래스: DNode
			- 클래스: LinkedBSTree
				데이터 삽입: InsertBST(_insertBST)
				데이터 삭제: DeleteBST(_deleteBST)
				데이터 검색: SearchBST(_searchBST)
				전체 원소 출력: printBSTAll(Preorder)
'''

# 클래스 설계: LinkedBSTree
class LinkedBSTree:
    class DNode:
        def __init__ (self, data, Llink=None, Rlink=None):
            self.data = data
            self.Llink = Llink
            self.Rlink = Rlink

    def __init__(self):
        self.__root = None

    # 소멸자: 너비 우선 순회
    def __del__(self):
        if not self.__root : return
			
    # 전체 노드 출력: 너비 우선 순회
    def printBSTAll(self) -> None:
        if not self.__root:
            print("\n데이터가 존재하지 않습니다.")
            return

    # 데이터 검색: 재귀적.비재귀적 용법
    def searchBST(self) -> None:
        # 데이터 검색: 재귀적 용법,  비재귀적 용법
        def _searchBST(root, data):
            if root == None or data == root.data:
                return root
            elif data < root.data :
                return _searchBST(root.Llink, data)
            else:
                return _searchBST(root.Rlink, data)

        if not self.__root:
            print("\n데이터가 존재하지 않습니다.")
            return
        
        print("\n이진 검색 트리: 데이터 검색")
        while True:
            num = int(input('임의의 정수 입력(종료: 0): '))
            if num == 0:
                break
            tNode = _searchBST(self.__root, num)
            if tNode:   print(f'{tNode.data} 키를 찾았습니다!!!')
            else:       print('키를 찾지 못했습니다.')

    # 데이터 삽입: 재귀적.비재귀적 용법
    def insertBST(self) -> None:
        # 데이터 삽입: 재귀적 용법, 비재귀적 용법
        def _insertBST(root, data):
            if root == None:
                root = self.DNode(data)
            elif data < root.data:
                root.Llink = _insertBST(root.Llink, data)
            else:
                root.Rlink = _insertBST(root.Rlink, data)
            return root

        print("\n이진 검색 트리: 데이터 삽입")
        while True:
            num = int(input('임의의 정수 입력(종료: 0): '))
            if num == 0:
                break
            self.__root = _insertBST(self.__root, num)

    # 데이터 삭제: 재귀적.비재귀적 용법
    def deleteBST(self) -> None:
        # 데이터 삭제: 비재귀적 용법
        def _deleteBST(root, data):
            if not self.__root:
                print("\n데이터가 존재하지 않습니다.")
                return
            pParent = None
            tNode = root

            while data != tNode.data:
                pParent = tNode
                if data < tNode.data:
                    tNode = tNode.Llink
                else:
                    tNode = tNode.Rlink

            if tNode.Llink == None and tNode.Rlink == None:     # 자식 노드가 없는 경우
                if pParent:
                    if pParent.Llink == tNode:
                        pParent.Llink = None
                    else:
                        pParent.Rlink = None
                else:
                    root = None

            # Case 2: 자식 노드가 1개 있는 경우
            elif tNode.Llink == None or tNode.Rlink == None:
                # pChild: 삭제할 노드의 자식 노드
                if tNode.Llink:
                    pChild= tNode.Llink
                else:
                    pChild = tNode.Rlink

                if pParent:
                    if pParent.Llink == tNode:
                        pParent.Llink = pChild
                    else:
                        pParent.Rlink = pChild
                else:
                    root = pChild

            # Case 3: 자식 노드가 둘 다 있는 경우
            elif tNode.Llink != None and tNode.Rlink != None:
                # 왼쪽 서브 트리에서 후계자(최댓값) 찾기
                sParent = tNode
                sNode = tNode.Llink
                while sNode.Rlink:
                    sParent = sNode
                    sNode = sNode.Rlink

                if sParent.Llink == sNode:
                    sParent.Llink = sNode.Llink
                else:
                    sParent.Rlink = sNode.Llink

                tNode.data = sNode.data
                tNode = sNode

            del tNode
            return root 

        print("\n이진 검색 트리: 데이터 삭제")
        while True:
            num = int(input('임의의 정수 입력(종료: 0): '))
            if num == 0:
                break
            self.__root = _deleteBST(self.__root, num)

if __name__ == '__main__':
    import os	# system
    import sys  # exit

    bst = LinkedBSTree()
    while True:
        os.system('clear')
        print('\n##### 이진 검색 트리 ###')
        print('1) 데이터 삽입')
        print('2) 데이터 삭제')
        print('3) 데이터 검색')
        print('4) 전체 출력')
        print('5) 프로그램 종료\n')
        choice = int(input('메뉴 선택: '))

        match choice:
            case 1: bst.insertBST()
            case 2: bst.deleteBST()
            case 3: bst.searchBST()
            case 4: bst.printBSTAll()
            case 5:
                sys.exit("\n프로그램 종료!!!")
            case _: print('\n잘못 선택 하셨습니다. \n')
        input("계속하려면 엔터키를 입력하십시오.")

    # del bst
    # bst.__del__
```
