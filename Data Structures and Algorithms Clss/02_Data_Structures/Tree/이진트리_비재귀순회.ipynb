```python
#깊이 우선 순회: 전위 순회
def Preorder(self)-> None:
    if not self.__root:
        return
    
    s = [self.__root]
    while s:
        tNode = s.pop()
        print(tNode.data, end= ' ')
        if tNode.Rlink: s.append(tNode.Rlink)
        if tNode.Llink: s.append(tNode.Llink)

# 깊이 우선 순회: 중위 순회
def Inorder(self):
    if not self.__root:
        return

    s = []
    tNode = self.__root
    while True:
        while tNode:
            s.append(tNode)
            tNode = tNode.Llink
        if not s:
            break

        tNode = s.pop()
        print(tNode.data, end = ' ')
        tNode = tNode.Rlink

# 깊이 우선 순회: 후위 순회
def Postorder(self):
    if not self.__root:
        return
    s = []
    tNode = self.__root
    while True:
        while tNode:
            s.append(tNode)
            tNode = tNode.Llink
        while s:
            pre = tNode
            tNode = s.pop()
            if tNode.Rlink:
                if tNode.Rlink == pre:
                    print(tNode.data, end = ' ')
                else:
                    s.append(tNode)
                    tNode = tNode.Rlink
                    break
            else: print(tNode.data, end = ' ')
        if not s:
            break
```

```python
# 이진 트리 구현 -- 연결 자료 구조

class LinkedBTree:
    class DNode:
        def __init__(self, data, Llink = None, Rlink = None):
            self.data = data
            self.Llink = Llink
            self.Rlink = Rlink

    def __init__(self):
        self.__root = None
    
    def __del__(self):
        if not self.__root:
            return
        q = [self.__root]
        while q:
            tNode = q.pop(0)
            if tNode.Llink:
                q.append(tNode.Llink)
            elif tNode.Rlink:
                q.append(tNode.Rlink)

            del tNode

    def Levelorder(self)-> None:
        if not self.__root:
            return
        q = [self.__root]
        while q:
            tNode = q.pop(0)
            print(f'{tNode.data}', end = ' ')
            if tNode.Llink : q.append(tNode.Llink)
            if tNode.Rlink : q.append(tNode.Rlink)

    def makeLinkedBTree(self, postfix)-> None:
        # 연산자 여부 판단
        def isOperator(op) -> bool:
            return op == '+' or op == '-' or op == '*' or op == '/'
        # 연산자와 피연산자 사이에 공백이 없는 경우를 위해...
        # split을 위해 연산자 앞 뒤에 공백을 넣어준다.
        postfix = postfix.replace('+', ' + ')
        postfix = postfix.replace('-', ' - ')
        postfix = postfix.replace('*', ' * ')
        postfix = postfix.replace('/', ' / ')
        postfix = postfix.split()

        s = []
        for ch in postfix:
            tNode = self.DNode(ch)
            if isOperator(ch):
                tNode.Rlink = s.pop()
                tNode.Llink = s.pop()
            s.append(tNode)
            
        self.__root = s.pop()

    #깊이 우선 순회: 전위 순회
    def Preorder(self)-> None:
        if not self.__root:
            return
        
        s = [self.__root]
        while s:
            tNode = s.pop()
            print(tNode.data, end= ' ')
            if tNode.Rlink: s.append(tNode.Rlink)
            if tNode.Llink: s.append(tNode.Llink)

    # 깊이 우선 순회: 중위 순회
    def Inorder(self):
        if not self.__root:
            return

        s = []
        tNode = self.__root
        while True:
            while tNode:
                s.append(tNode)
                tNode = tNode.Llink
            if not s:
                break

            tNode = s.pop()
            print(tNode.data, end = ' ')
            tNode = tNode.Rlink

    # 깊이 우선 순회: 후위 순회
    def Postorder(self):
        if not self.__root:
            return
        s = []
        tNode = self.__root
        while True:
            while tNode:
                s.append(tNode)
                tNode = tNode.Llink
            while s:
                pre = tNode
                tNode = s.pop()
                if tNode.Rlink:
                    if tNode.Rlink == pre:
                        print(tNode.data, end = ' ')
                    else:
                        s.append(tNode)
                        tNode = tNode.Rlink
                        break
                else: print(tNode.data, end = ' ')
            if not s:
                break

if __name__ == '__main__':
    postfix = input('트리를 구성할 후위 수식')
    BTree = LinkedBTree()
    BTree.makeLinkedBTree(postfix)

    print(f'트리를 구성할 후위 수식 : {postfix}')

    print(f'Preorder    : ', end= ' ')
    BTree.Preorder()
    print()

    print(f'Inorder     : ', end = ' ')
    BTree.Inorder()
    print()

    print(f'Postorder   : ', end = ' ')
    BTree.Postorder()
    print()

    print(f'Levelorder  : ', end = ' ')
    {BTree.Levelorder()}
    print()
```
