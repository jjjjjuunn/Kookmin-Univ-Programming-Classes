```python
# 이진 트리 구현 -- 연결 자료 구조

class LinkedBTree:
    class DNode:
        def __init__(self, data, Llink = None, Rlink = None):
            self.data = data
            self.Llink = Llink
            self.Rlink = Rlink

    def __init__(self):
        self.__root = None
    
    def __del__(self):
        if not self.__root:
            return
        q = [self.__root]
        while q:
            tNode = q.pop(0)
            if tNode.Llink:
                q.append(tNode.Llink)
            elif tNode.Rlink:
                q.append(tNode.Rlink)

            del tNode

    def Levelorder(self)-> None:
        if not self.__root:
            return
        q = [self.__root]
        while q:
            tNode = q.pop(0)
            print(f'{tNode.data}', end = ' ')
            if tNode.Llink : q.append(tNode.Llink)
            if tNode.Rlink : q.append(tNode.Rlink)

    def makeLinkedBTree(self, postfix)-> None:
        # 연산자 여부 판단
        def isOperator(op) -> bool:
            return op == '+' or op == '-' or op == '*' or op == '/'
        # 연산자와 피연산자 사이에 공백이 없는 경우를 위해...
        # split을 위해 연산자 앞 뒤에 공백을 넣어준다.
        postfix = postfix.replace('+', ' + ')
        postfix = postfix.replace('-', ' - ')
        postfix = postfix.replace('*', ' * ')
        postfix = postfix.replace('/', ' / ')
        postfix = postfix.split()

        s = []
        for ch in postfix:
            tNode = self.DNode(ch)
            if isOperator(ch):
                tNode.Rlink = s.pop()
                tNode.Llink = s.pop()
            s.append(tNode)
            
        self.__root = s.pop()

    def Preorder(self)-> None:
        def _Preorder(tNode):
            if tNode:
                print(tNode.data, end=' ')
                _Preorder(tNode.Llink)
                _Preorder(tNode.Rlink)
        _Preorder(self.__root)
        
    def Inorder(self)-> None:
        def _Inorder(tNode):
            if tNode:
                _Inorder(tNode.Llink)
                print(tNode.data, end=' ')
                _Inorder(tNode.Rlink)
        _Inorder(self.__root)

    def Postorder(self)-> None:
        def _Postorder(tNode):
            if tNode:
                _Postorder(tNode.Llink)
                _Postorder(tNode.Rlink)
                print(tNode.data, end=' ')
        _Postorder(self.__root)

if __name__ == '__main__':
    postfix = input('트리를 구성할 후위 수식')
    BTree = LinkedBTree()
    BTree.makeLinkedBTree(postfix)

    print(f'트리를 구성할 후위 수식 : {postfix}')

    print(f'Preorder    : ', end= ' ')
    BTree.Preorder()
    print()

    print(f'Inorder     : ', end = ' ')
    BTree.Inorder()
    print()

    print(f'Postorder   : ', end = ' ')
    BTree.Postorder()
    print()

    print(f'Levelorder  : ', end = ' ')
    {BTree.Levelorder()}
    print()
```
