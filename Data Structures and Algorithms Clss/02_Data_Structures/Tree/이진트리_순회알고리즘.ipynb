```python
# ì´ì§„ íŠ¸ë¦¬ êµ¬í˜„ -- ì—°ê²° ìë£Œ êµ¬ì¡°

class LinkedBTree:
    class DNode:
        def __init__(self, data, Llink = None, Rlink = None):
            self.data = data
            self.Llink = Llink
            self.Rlink = Rlink

    def __init__(self):
        self.__root = None
    
    def __del__(self):
        if not self.__root:
            return
        q = [self.__root]
        while q:
            tNode = q.pop(0)
            if tNode.Llink:
                q.append(tNode.Llink)
            elif tNode.Rlink:
                q.append(tNode.Rlink)

            del tNode

    def Levelorder(self)-> None:
        if not self.__root:
            return
        q = [self.__root]
        while q:
            tNode = q.pop(0)
            print(f'{tNode.data}', end = ' ')
            if tNode.Llink : q.append(tNode.Llink)
            if tNode.Rlink : q.append(tNode.Rlink)

    def makeLinkedBTree(self, postfix)-> None:
        # ì—°ì‚°ì ì—¬ë¶€ íŒë‹¨
        def isOperator(op) -> bool:
            return op == '+' or op == '-' or op == '*' or op == '/'
        # ì—°ì‚°ìì™€ í”¼ì—°ì‚°ì ì‚¬ì´ì— ê³µë°±ì´ ì—†ëŠ” ê²½ìš°ë¥¼ ìœ„í•´...
        # splitì„ ìœ„í•´ ì—°ì‚°ì ì• ë’¤ì— ê³µë°±ì„ ë„£ì–´ì¤€ë‹¤.
        postfix = postfix.replace('+', ' + ')
        postfix = postfix.replace('-', ' - ')
        postfix = postfix.replace('*', ' * ')
        postfix = postfix.replace('/', ' / ')
        postfix = postfix.split()

        s = []
        for ch in postfix:
            tNode = self.DNode(ch)
            if isOperator(ch):
                tNode.Rlink = s.pop()
                tNode.Llink = s.pop()
            s.append(tNode)
            
        self.__root = s.pop()

    def Preorder(self)-> None:
        def _Preorder(tNode):
            if tNode:
                print(tNode.data, end=' ')
                _Preorder(tNode.Llink)
                _Preorder(tNode.Rlink)
        _Preorder(self.__root)
        
    def Inorder(self)-> None:
        def _Inorder(tNode):
            if tNode:
                _Inorder(tNode.Llink)
                print(tNode.data, end=' ')
                _Inorder(tNode.Rlink)
        _Inorder(self.__root)

    def Postorder(self)-> None:
        def _Postorder(tNode):
            if tNode:
                _Postorder(tNode.Llink)
                _Postorder(tNode.Rlink)
                print(tNode.data, end=' ')
        _Postorder(self.__root)

if __name__ == '__main__':
    postfix = input('íŠ¸ë¦¬ë¥¼ êµ¬ì„±í•  í›„ìœ„ ìˆ˜ì‹')
    BTree = LinkedBTree()
    BTree.makeLinkedBTree(postfix)

    print(f'íŠ¸ë¦¬ë¥¼ êµ¬ì„±í•  í›„ìœ„ ìˆ˜ì‹ : {postfix}')

    print(f'Preorder    : ', end= ' ')
    BTree.Preorder()
    print()

    print(f'Inorder     : ', end = ' ')
    BTree.Inorder()
    print()

    print(f'Postorder   : ', end = ' ')
    BTree.Postorder()
    print()

    print(f'Levelorder  : ', end = ' ')
    {BTree.Levelorder()}
    print()
```
