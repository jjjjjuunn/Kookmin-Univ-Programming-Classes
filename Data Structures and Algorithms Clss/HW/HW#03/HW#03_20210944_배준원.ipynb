## 연습문제 #01
### 데이터 처리: 단순 연결 리스트
**임의의 정수값을 입력 받아 출력하는 프로그램을 작성하세요.**

- 단, 원소 삽입은 항상 마지막 노드로 삽입된다고 가정한다.
- 또한 전체 원소에 대하여 삭제( 메모리반납 )후 프로그램을 종료한다.

```python
# class
# tail과 count가 있는 버전

class SNode:
    def __init__(self, data):
        self.__data = data
        self.__link = None
    def getData(self):
        return self.__data
    def getLink(self):
        return self.__link
    def setData(self, data):
        self.__data = data
    def setLink(self, link):
        self.__link = link
    
class SLinkedList():
    def __init__(self):
        self.__head = None
        self.__tail = None
        self.__count = 0

    def __del__(self):
        while self.__head:
            old = self.__head
            self.__head = old.getLink()
            old.setLink(None)
            del old
        self.__tail = None
        self.__count = 0
        
    def isEmpty(self) -> bool:
        return self.__head is None
    def countNode(self) -> int:
        return self.__count
    def frontNode(self) -> SNode:
        return self.__head
    def rearNode(self) ->SNode:
        return self.__tail

    def addRear(self, num) -> None:
        newNode = SNode(num)
        if self.isEmpty():
            self.__head = newNode
            self.__tail = newNode
        else:
            self.__tail.setLink(newNode)
            self.__tail = newNode
        
        self.__count += 1

    def removeFront(self) -> int:
        if self.isEmpty(): return
        else:
            old = self.__head
            self.__head = old.getLink()
            if self.__head == None:
                self.__tail = None
            old_data = old.getData()
            old.setLink(None)
            del old
            self.__count -= 1
            return old_data

    def printLinkedList(self) -> None:
        if self.isEmpty():
            print("이 단순연결 리스트는 비어있습니다 !!!")
            return
        else:
            print("### 입력된 데이터 ###")
            node = self.__head
            while node:
                end = ' ->> ' if node.getLink() else '\n'
                print(f'{node.getData()}', end = end)
                node = node.getLink()
```

```python
# main
sList = SLinkedList()

while True:
    num = int(input('임의의 정수 입력(종료: 0): '))
    if num == 0:
        break
    else:
        sList.addRear(num)

sList.printLinkedList()

print(f'노드의 총 개수: {sList.countNode()}')

front = sList.frontNode()
rear = sList.rearNode()

if front is None:
    print(f'입력된 데이터가 없습니다 !')
else:
    print(f'첫번째 노드의 데이터: {front.getData()}')
    if rear is not None:
        print(f'마지막 노드의 데이터: {rear.getData()}')

del sList
```

## 연습문제 #03
### 스택 구현: 단순 연결 리스트

```python
# class 
# stack : LIFO( Last Input First Output )
class LinkedStack:
    class SNode:
        def __init__(self, data, link = None):
            self.data = data
            self.link = link


    def __init__(self):
        self.__top = None # head
        self.__count = 0

    def __del__(self):
        while self.__top:
            old = self.__top
            self.__top = old.link
            old.link = None
            del old
            
        self.__count = 0

    def empty(self)-> bool:
        return self.__top is None
    
    def size(self)-> int:
        return self.__count

    def push(self, data)-> None:
        newStack = self.SNode(data, self.__top)
        self.__top = newStack
        self.__count += 1

    def pop(self)-> None:
        if self.empty():
            print('더이상 삭제할 데이터가 없습니다.')
            return None
        
        old = self.__top
        self.__top = old.link
        old.link = None
        print(f'삭제된 데이터: {old.data}')
        del old
        self.__count -= 1

    def top(self):
        if self.empty(): return None
        else:
            return self.__top.data
    
    def printStack(self)-> None:
        if self.empty():
            print("입력된 데이터가 없습니다.")
        else:
            print('### 입력된 데이터 ###')
            node = self.__top
            elements = []
            while node:
                elements.append(str(node.data))
                node = node.link
            print(f'STACK [{' '.join(elements)}]')
```

```python
# main
from IPython.display import clear_output

stack_linked = LinkedStack()

while True:
    print('### 스택 구현: 단순연결리스트 ###')
    print('1) 데이터 삽입: push')
    print('2) 데이터 삭제: pop')
    print('3) 전체 출력')
    print('4) 프로그램 종료\n')

    menu_num = int(input('메뉴 선택: '))

    match menu_num:
        case 1:
            while True:
                data = int(input('삽입 할 데이터 입력(종료: 0): '))
                if data == 0:
                    break
                stack_linked.push(data)
        case 2:
            stack_linked.pop()
        case 3:
            stack_linked.printStack()

        case 4:
            del stack_linked
            break
    
    input('계속하려면 enter 버튼을 입력하십시오')
    clear_output(wait = True)
    
```

## 연습문제 #04
### 큐 구현: 단순 연결 리스트

```python
# class : FIFO(First Input First Output)
class LinkedQueue:
    class SNode:
        def __init__(self, data, link = None):
            self.data = data
            self.link = link
    def __init__(self):
        self.__front = None
        self.__rear = None
        self.__count = 0

    def __del__(self):
        while self.__front:
            old = self.__front
            self.__front = old.link
            old.link = None
            del old

        self.__rear = None
        self.__count = 0 

    def empty(self)-> bool:
        return self.__front is None
    
    def size(self)-> int:
        return self.__count
    
    def push(self, data)-> None:
        NewQueue = self.SNode(data)
        if self.empty():
            self.__front = NewQueue
            self.__rear = NewQueue
        else:
            self.__rear.link = NewQueue
            self.__rear = NewQueue

        self.__count += 1

    def pop(self):
        if self.empty():
            print('더이상 제거할 데이터가 없습니다.')
            return None
        else:
            old = self.__front
            self.__front = old.link
            if self.__front is None:
                self.__rear = None
            old.link = None
            val = old.data
            print(f'삭제된 데이터: {val}')
            del old
            self.__count -= 1
            return val

    def front(self):
        if self.empty():
            return None
        else:
            return self.__front.data
        
    def back(self):
        if self.empty():
            return None
        else:
            return self.__rear.data
        
    def printQueue(self):
        if self.empty():
            print('데이터가 없습니다.')

        else:
            print('#### 입력된 데이터 ####')
            node = self.__front
            elements = []
            while node:
                elements.append(str(node.data))
                node = node.link
            print(f'QUEUE [ {" ".join(elements)} ]')
```

```python
# main
from IPython.display import clear_output

Queue_Linked = LinkedQueue()

while True:
    print('### 스택 구현: 단순연결리스트 ###')
    print('1) 데이터 삽입: push')
    print('2) 데이터 삭제: pop')
    print('3) 전체 출력')
    print('4) 프로그램 종료\n')

    menu_num = int(input('메뉴 선택: '))

    match menu_num:
        case 1:
            while True:
                data = int(input('삽입 할 데이터 입력(종료: 0): '))
                if data == 0:
                    break
                Queue_Linked.push(data)
        case 2:
            Queue_Linked.pop()
        case 3:
            Queue_Linked.printQueue()

        case 4:
            del Queue_Linked
            break

        case 5:
            print(Queue_Linked.size())
    
    input('계속하려면 enter 버튼을 입력하십시오')
    clear_output(wait = True)
```
