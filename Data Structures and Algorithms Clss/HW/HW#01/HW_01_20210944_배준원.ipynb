```python
# 과제 #01번
x, y, z = input('수식을 입력하세요(예: 10 + 20): ').split()
x = int(x)
z = int(z)
res = 0
error_occurred = False

match y:
    case '+': res = x + z
    case '-': res = x - z
    case '*': res = x * z
    case '/':
        try: 
            res = x / z
        except ZeroDivisionError as e : 
            print(f'오류: {e}') 
            error_occurred = True
        
    case default:
        print("오류: 지원하지 않는 연산자입니다.")
        error_occurred = True

if not error_occurred:
    print(f'{x} {y} {z} = {res}')

```

## 과제 #01 추가설명

### 2
match - case 문을 사용해 y에 따라 수행할 연산을 분리했다.
+, -, *, / 외의 연산자가 입력될 경우 오류를 멘트를 발생시키기 위해 default 또한 추가했다.
나눗셈 연산에서 0으로 나누는 오류를 잡기 위해 try - except를 이용하여 예외 처리를 실행하였다. 

### 3
오류가 발생될 경우 마지막 결과물을 출력하는 것을 방지하기 위해 error_occurred라는  플래그 변수를 통해 에러가 발생하지 않았을 때만 출력하도록 하였다.


```python
# 과제 #02번 : 최댓값과 최솟값 반환 함수
def MAXMIN(sList : list) -> int:
    Max_val, Min_val = sList[0], sList[0]
    for i in sList:
        if i > Max_val:
            Max_val = i
        if i < Min_val:
            Min_val = i
    return Max_val, Min_val

if __name__ == '__main__':
    print("임의의 정수를 입력하세요...")
    i = 0
    sList = []
    while True:
        num = input(f"sList[{i}]: ")
        if num == 'end' or num == '':
            break
        sList.append(int(num))
        i += 1

    result = MAXMIN(sList)
    print(f"최대값: {result[0]}, 최소값 {result[1]}")

```

```python
# 과제 #03: 전화번호부 : 딕셔너리
import os

Phone_num_name_dict = {}

def input_mode():
    print("[ 입력 모드 ]")
    while True:
        name = input("이름(입력 종료: end): ")
        if name == "end" or name == '':
            input("계속하려면 엔터를 눌러주십시오...")
            return 0
        phone_num = input("연락처: ")

        Phone_num_name_dict.setdefault(name, phone_num)

def searching_mode():
    print("[ 검색 모드 ]")
    while True:
        name = input("이름(검색 종료: end): ")
        if name == "end" or name == "":
            input("계속하려면 엔터를 눌러주십시오...")
            return 0
        print(f"{name}의 전화번호는 {Phone_num_name_dict.get(name)}입니다.")

def output_mode():
    print("##### 전화번호부 #####")
    for item in Phone_num_name_dict.items():
        print(item)
    input("계속하려면 엔터를 눌러주십시오...")
    return 0

if __name__ == '__main__':
    while True:
        os.system('clear')
        print("##### 전화번호부 #####")
        print("1: 입력")
        print("2: 검색")
        print("3: 전체 출력")
        print("4: 프로그램 종료")
        menu = int(input("메뉴 선택: "))

        match menu:
            case 1:
                input_mode()
            case 2:
                searching_mode()
            case 3:
                output_mode()
            case 4:
                print("프로그램 종료 !!!")
                break
            case default:
                print("오류: 지원하지 않는 메뉴입니다. 다시 입력해주십시오")
```

```python
# 과제 #04: 로또 복권: 번호 자동 생성
import random

num = int(input("로또 복권을 몇 장 구입하시겠습니까?: "))

for _ in range(num):
    lotto_total = []
    while len(lotto_total) < 6:
        lotto_num = random.randint(1,45)
        if lotto_num not in lotto_total:
            lotto_total.append(lotto_num)
    lotto_total.sort()
    print(lotto_total)
```

```python
# 과제 05 : 계산기 클래스 -> 상속과 메소드 재정의
class Cal:
    def __init__(self, first = 0, second = 0):
        self.__first = first
        self.__second = second

    def add(self):
        return self.__first + self.__second

    def sub(self):
        return self.__first - self.__second

    def mul(self):
        return self.__first * self.__second

    def div(self):
        return self.__first / self.__second

    def getnum(self):
        return self.__first, self.__second

class NewCal(Cal):
    def div(self):
        try:
           return self.getnum()[0] / self.getnum()[1]
        
        except ZeroDivisionError as e : 
            print(f'오류 : {e}')
            return 0

    def pow(self):
        return self.getnum()[0] ** self.getnum()[1]


if __name__ == '__main__':
    print("두 개의 정수를 입력하세요...")
    first = int(input("first : "))
    second = int(input("second : "))

    Calculate = NewCal(first, second)
    
    print(f'{first} + {second} = {Calculate.add()}')
    print(f'{first} - {second} = {Calculate.sub()}')
    print(f'{first} * {second} = {Calculate.mul()}')
    print(f'{first} / {second} = {Calculate.div()}')
    print(f'{first} ** {second} = {Calculate.pow()}')

```
